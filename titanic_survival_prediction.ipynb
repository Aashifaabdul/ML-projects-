```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
file_path = '/content/tested.csv'  # Replace with your file path
df = pd.read_csv(file_path)
df.head()
```

```python
print("\nMissing values in the DataFrame:")
print(df.isna())
print(df.isna().sum())
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
print(df.isna().sum())
```

```python
df['Fare'] = df['Fare'].fillna(df['Fare'].mode()[0])
df = df.drop(columns='Cabin')
print(df.isna().sum())
```

```python
df
```

```python
df = pd.get_dummies(df, columns=['Sex', 'Embarked'], drop_first=True)
```

```python
df
```

```python
X = df[['PassengerId', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Sex_male', 'Embarked_Q', 'Embarked_S']]
y = df['Survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
print("\nTraining data shape:", X_train.shape)
print("Testing data shape:", X_test.shape)
```

```python
model = LogisticRegression(max_iter=30)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print("\nPredicted values for the test set:")
print(y_pred)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("\nAccuracy of the model:", accuracy)
```
